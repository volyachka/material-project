```python
import numpy as np
import pandas as pd
```

```python
mp_ids = pd.read_csv('fv.v2.Kahle2020-upd2024-03-04/exported.predictions.mp.v2.csv')['material_id'].to_numpy()
```

```python
def get_predictions(preds):
    agg_preds = np.zeros(len(mp_ids))

    for prediction in preds:
        agg_preds += prediction

    agg_preds /= len(preds)

    results = pd.DataFrame({'ids': mp_ids[np.argsort(agg_preds)[::-1]][:10], 
              'probs': agg_preds[np.argsort(agg_preds)[::-1]][:10]})
    
    return results
```

```python
lightgbm_only_featurizers = np.loadtxt('predictions/lightgbm_only_featurizers_max_depth_3_lr_0.5_n_estimators_3.txt')

top_lightgbm_only_featurizers = get_predictions(lightgbm_only_featurizers)

top_lightgbm_only_featurizers
```

```python
lightgbm_only_nn = np.loadtxt('predictions/lightgbm_only_nn_max_depth_3_lr_0.9_n_estimators_3.txt')

top_lightgbm_only_nn = get_predictions(lightgbm_only_nn)

top_lightgbm_only_nn
```

```python
lightgbm_merged = np.loadtxt('predictions/lightgbm_merged_max_depth_3_lr_0.5_n_estimators_3.txt')

top_lightgbm_merged = get_predictions(lightgbm_merged)

top_lightgbm_merged
```

```python
catboost_only_featurizers = np.loadtxt('predictions/catboost_only_featurizers_max_depth_3_lr_0.5_n_estimators_3.txt')

top_catboost_only_featurizers = get_predictions(catboost_only_featurizers)

top_catboost_only_featurizers
```

```python
catboost_only_featurizers = np.loadtxt('predictions/catboost_only_nn_max_depth_3_lr_0.9_n_estimators_3.txt')

top_catboost_only_featurizers = get_predictions(catboost_only_featurizers)

top_catboost_only_featurizers
```
